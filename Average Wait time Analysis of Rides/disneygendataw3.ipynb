```python
import pandas as pd
import random
from datetime import datetime, timedelta

# Function to generate random date and time
def generate_random_datetime():
    start_date = datetime(2023, 1, 1)
    end_date = datetime(2023, 12, 31)
    random_date = start_date + timedelta(days=random.randint(0, (end_date - start_date).days))
    random_time = timedelta(hours=random.randint(8, 21), minutes=random.randint(0, 59))
    return random_date + random_time

# Function to generate random data for the specified attractions
def generate_random_data(attraction, ride_name, avg_wait_time):
    date = generate_random_datetime().strftime('%Y-%m-%d')
    time = generate_random_datetime().strftime('%I:%M %p')
    total_tickets_sold = random.randint(50, 200)
    revenue = total_tickets_sold * random.uniform(8, 15)
    ticket_type = random.choice(['Regular', 'FastPass', 'VIP'])
    park_area = attraction
    scheduled_time = generate_random_datetime().strftime('%I:%M %p')
    average_wait_time = avg_wait_time  # Use the specified average wait time
    wait_time = random.randint(5, average_wait_time) if not is_peak_hour(time) else random.randint(20, 60)
    ride_capacity = random.randint(100, 200)
    fastpass_reservations = random.randint(20, 50)
    age_group = random.choice(['Child', 'Adult'])
    total_visits = random.randint(1, 10)
    average_rating = round(random.uniform(3.5, 5.0), 1)

    return [date, time, total_tickets_sold, revenue, ticket_type, park_area, ride_name,
            scheduled_time, average_wait_time, wait_time, ride_capacity,
            fastpass_reservations, age_group, total_visits, average_rating]

# Function to check if it's peak hours
def is_peak_hour(time):
    return ((time >= '10:00 AM' and time <= '11:30 AM') or
            (time >= '01:00 PM' and time <= '02:00 PM'))

# List of attractions and rides with specified average wait times
attractions = {
    'Frontierland': {'Splash Mountain': 81, 'Big Thunder Mountain': 69, 'Railroad': 21},
    'Fantasyland': {'Enchanted Tales with Belle': 61, 'Seven Dwarfs Mine Train': 119, "Peter Pan's Flight": 88},
    'Tomorrowland': {'Space Mountain': 92, "Buzz Lightyear's Space Ranger Spin": 62, 'Tomorrowland Speedway': 46}
    # ... (rest of the attractions and rides)
}

# Generate random data for each ride with specified average wait times
data = []
while len(data) < 5000:
    for park_area, rides in attractions.items():
        for ride_name, avg_wait_time in rides.items():
            data.append(generate_random_data(park_area, ride_name, avg_wait_time))

# Trim excess records to achieve exactly 5000 records
data = data[:5000]

# Create a DataFrame
columns = ['Date', 'Time', 'TotalTicketsSold', 'Revenue', 'TicketType', 'ParkArea', 'RideName',
           'ScheduledTime', 'AverageWaitTime', 'WaitTime', 'RideCapacity', 'FastPassReservations',
           'AgeGroup', 'TotalVisits', 'AverageRating']

df = pd.DataFrame(data, columns=columns)

# Save the DataFrame to a CSV file
df.to_csv('theme_park_dataset_updated_5000.csv', index=False)
```
